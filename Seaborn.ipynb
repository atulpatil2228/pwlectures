```python
import pandas as  pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset("tips")
```

```python
df
```

```python
##numerical data plotting  #This function provides access to several different axes-level functions
##that show the relationship between two variables with semantic mappings
##of subsets. The kin-d parameter selects the underlying axes-level
## function to use:
sns.relplot(x= "total_bill" , y = "tip", data  =  df)
```

```python
sns.scatterplot(x= df.total_bill , y = df.tip ,data = df , hue = 'smoker') ## by default it gives scatter plot, kind = line gives you line plot
```

```python
## as the bill increase , the tip also increases
## highest tip was paid by a smoker
sns.relplot(x= "total_bill" , y = "tip", data  =  df , hue = 'smoker')
```

```python
sns.relplot(x= "total_bill" , y = "tip", data  =  df , hue = 'size')
```

```python
sns.relplot(x= "total_bill" , y = "tip", data  =  df ,style = 'smoker', hue = 'time')## hue gives different colour
```

```python
sns.relplot(x= "total_bill" , y = "tip", data  =  df ,style = 'size', )
```
