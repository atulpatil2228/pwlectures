```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df.mean()
```

```python
df.mean(numeric_only=True)
```

```python
df.median(numeric_only=True)
```

```python
df.std(numeric_only=True)
```

```python
df.Age.describe()
```

```python
df
```

```python
## what is the average fair paid by people who survivde?
df[df.Survived == 1]['Fare'].mean()
```

```python
## avg fair who didnt survived
df[df.Survived == 0]['Fare'].mean()
```

```python
## groupby

df.groupby("Survived").mean(numeric_only=True)
```

```python
df.groupby("Survived").min(numeric_only=True)
```

```python
df.groupby("Survived").sum(numeric_only=True)
```

```python
df.groupby('Survived').describe()
```

```python
import numpy as np

df.groupby(['Survived'])['Fare'].agg([min,'max','median','count',np.std,'var'])
```
