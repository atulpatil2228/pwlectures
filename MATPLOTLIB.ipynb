```python

x = [1,2,3,4,5]
y1 = [5,1,7,8,2]
y2 = [7,8,3,5,1]
plt.plot(x,y1,color = "red",marker = "o",linestyle = "--",linewidth =4,markersize =11,label = "liney1")
plt.plot(x,y2,color = "b",marker = "o",linestyle = "--",linewidth =4,markersize =11,label ="liney2")
plt.xlabel("x data axis")
plt.ylabel("y data axix")

plt.title("Multiples line in one plot")
plt.legend()
plt.show()
```

```python
## pie chart >> to uderstand the % distribution of data


langs = ['C','Python','Java']

students = [20,100,40]

plt.pie(students,labels = langs,autopct ='%1.1f%%')
plt.title("Pie chart")
plt.show()
```

```python
explode  = (0,0.2,0.4)

langs = ['C','Python','Java']

students = [20,100,40]

plt.pie(students,labels = langs,autopct ='%1.1f%%', explode = explode,shadow = True)
plt.title("Pie chart")
plt.show()
```

```python
## 3d plot

x =  np.random.rand(20)
y = np.random.rand(20)
z = np.random.rand(20)

fig = plt.figure()
ax = fig.add_subplot(projection = "3d")
ax.scatter(x,y,z,color = "red")
```
